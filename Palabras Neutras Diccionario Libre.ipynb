```python
import pandas as pd
```

```python
dicc = pd.read_csv('palabras.csv',low_memory=False)
```

```python
dicc.shape
```

```python
dicc.sample(n=20)
```

```python
tipos_de_palabras = dicc.tipo.dropna().unique()
```

```python
len(tipos_de_palabras)
```

```python
tipos_verbos_neutros = {t for t in tipos_de_palabras if ('verbo' in t) and ('participio' not in t)}
tipos_neutros = [
    'adverbio',
    'nombre/ambiguo',
    'nombre/común',
] + list(tipos_verbos_neutros)
```

```python
len(tipos_neutros)
```

```python
palabras_neutras = dicc[dicc['tipo'].isin(tipos_neutros) & dicc['género'].isnull()]
```

```python
def replace_name(name):
    if 'verbo' in name:
        name = 'verbo'
    return name
```

```python
palabras_neutras['tipo'] = palabras_neutras['tipo'].apply(replace_name)
```

```python
(palabras_neutras.tipo.unique())
```

```python
origen = dicc['origen'].unique()
```

```python
palabras_neutras[palabras_neutras['tipo'] == 'verbo']
```

```python
groups = palabras_neutras.groupby('tipo')
```

```python
for name,group_df in groups:
    name = (name.replace('/','_'))
    group_df['palabra'].to_csv(f'palabras_neutras/dicc_libre/{name}',columns=['palabra'],index=False)
```
